```python
import os
from PIL import Image
```

```python
def delete_dot_ds_store(path):
    for root, dirnames, filenames in os.walk(path):
        for filename in filenames:
            if filename == '.DS_Store':
                os.remove(os.path.join(root, filename))
                
```

```python
raw_datadir = "/Users/zhaonan/UAL-CCI/07-MScAdvancedProject/raw_dataset/6-UGATIT-face/trainB_origin"
data_path = os.path.join(raw_datadir)
delete_dot_ds_store(data_path) 
data_list=os.listdir(data_path)
data_list
```

```python
ind = 0
for i in data_list:
    original_img_name = os.path.splitext(os.path.basename(os.path.join(data_path,i)))[0]
    filename = os.path.basename(os.path.join(data_path,i)) # with extension
    os.rename(raw_datadir+'/'+filename, raw_datadir+'/'+str(ind)+'.jpg') # 
    ind = ind + 1
    print(raw_datadir+'/'+str(ind))
```
